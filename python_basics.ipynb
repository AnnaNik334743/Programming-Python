Общий приоритет операторов.

Оператор возведения в степень исключение из этого правила. Из двух операторов ** сначала выполнится правый, а потом левый.

1. Скобки, то есть выражения в них `()`
2. Возведение в степень `**`
3. Унарные плюс, минус и битовое отрицание `+x`, `-x`, `~x`
4. Умножение, деление, целочисленное деление, остаток от деления `*`, `/`, `//`, `%`
5. Сложение и вычитание `+`, `-`
6. Битовые сдвиги `<<`, `>>`
7. Битовое И `&`
8. Битовое исключающее ИЛИ (XOR) `^`
9. Битовое ИЛИ `|`

11. Логическое НЕ `not`
12. Логическое И `and`

## Условия

Условные операторы разветвляют линейную структуру программы. В Python условия можно проверять с помощью операторов if, elif и else. Общая структура:

```python
  if условие1:
    действия, когда условие1 верно
  elif условие2:
      действия, когда условие1 неверно, а условие2 верно
  elif условие3:
      действия, когда условие_1 условие_2 неверны, а условие_3 верно
  ...  # Сколько угодно elif-ов
  else:
      действия, когда все условия неверны
```

````python
x = 10
if x > 5:
    print("x больше 5")
````

````python
x = 3
if x > 5:
    print("x больше 5")
else:
    print("x меньше или равно 5")
````

````python
x = 10
if x > 10:
    print("x больше 10")
elif x == 10:
    print("x равно 10")
else:
    print("x меньше 10")
````

Начиная с версии Python 3.10, введен новый оператор match/case для сопоставления значений с шаблонами.

```python
match значение:
    case шаблон1:
        # Код, выполняющийся, если значение соответствует шаблону1
    case шаблон2:
        # Код, выполняющийся, если значение соответствует шаблону2
    ...
    case шаблонN:
        # Код, выполняющийся, если значение соответствует шаблонуN
    case _:
        # Код, выполняющийся, если значение не соответствует ни одному из шаблонов
```

````python
state = input()

match state:
    case "кирпич мокрый":
      print("идет дождь")
    case "кирпич сухой":
      print("дождя нет")
    case "кирпич отбрасывает тень":
      print("ясно")
    case "кирпича не видно":
      print("туман")
    case "кирпич качается":
      print("ветрено")
    case "на кирпиче что-то белое":
      print("снег")
    case "кирпич ходит вверх-вниз":
      print("землетрясение")
    case "кирпича нет":
      print("обновление данных")
    case _:
      print("я не знаю, что происходит, но БЕГИТЕ оттуда")
````

## Циклы

Цикл нужен для многократного повторения одних и тех же операций. В большинстве языков программирования есть три типа циклов: с предусловием (while), с счетчиком (for) и с постусловием (until). В Python есть только первые два.


Цикл с предусловием. Код в теле цикла будет выполняться, пока условие верно.

```python
while условие:
    код, который выполянется в цикле
else:
    код, который выполнится разово, как только мы выйдем из цикла
```

Напишем программу, которая будет считать сумму всех введеннных пользователем чисел, если они делятся на два без остатка. Когда пользователь введет 0, программа завершится.

````python
s = 0
n = int(input())
while n != 0:
  if n % 2 == 0:
    s += n
  n = int(input())
print(s)
````

Чуть-чуть поменяем условие. Теперь вводить можно только числа, большие нуля. Обратите внимание на блок `else` после тела цикла. Когда мы выйдем из цикла, то зайдем в него.

````python
s = 0
n = int(input())
while n > 0:
  if n % 2 == 0:
    s += n
  n = int(input())
else:
  print("Отрицательное число!")
print(s)
````

Теперь поговорим про цикл с счетчиком (цикл for).

В некоторых языках программирования он строится по следующему принципу: берем переменную-счетчик, устанавливаем для неё начальное значение, конечное и шаг. Похоже на функцию range(), не правда ли?

Но цикл в Питоне работает несколько иначе, по принципу

```python
for переменная in итерируемый объект:
    тело цикла
```

Что такое итерируемый объект? Это любая последовательность. Например, это может быть набор переменных, список, состоящая из символов строка, заданная функцией range() последовательность и т.п. Переменная-счетчик на самом деле не является счетчиком! В ходе выполнения цикла меняется не она, а автоматически создающийся "указатель" в итерируемом объекте. И переменная всего лишь принимает то значение, на котором на данной итерации цикла оказался указатель. Переменную можно менять в теле цикла, в Python никаких санкций из-за этого не будет.

````python
for i in range(4):
  print(i)
````

````python
for j in range(5):
  print(j if j % 3 == 0 else None)
  j = 100   # C++ или Pascal стали бы ругаться на это. Питону все равно, потому что j - это не счетчик
````

````python
for letter in "JhgHghsatgGCgfcofeioi":
  if letter in "eioa":  # Проверяем, содержится ли текущий символ в указанной последовательности
    print(letter)
````

При работе с циклами необходимо знать об операторах `break` и `continue`. `break` прерывает выполнение всего цикла, а `continue` пропускает текущую итерацию. Оба оператора работают только с циклом текущего уровня.

Напишем код, задача которого - выяснить для всех положительных введенных пользователем чисел, являются ли они простыми (то есть, имеют ли в качестве делителя только единицу и само себя). Когда введен 0, программа завершается.

````python
n = 1
while n != 0:
  n = int(input())
  if n < 0:
    print("Number must be greater than 0")
    continue  # Относится к циклу while
  elif n == 0:
    break  # Относится к циклу while
  else:
    counter = 0
    for i in range(2, int(n**0.5 + 1)):  # Вложенный цикл - проверка числа на простоту
      if n % i == 0:
        break  # Относится к циклу for
    else:  # Обратите внимание: в блок else цикла можно перейти только когда цикл завершился естественным образом без break
      print("Yes")
      continue  # Относится к циклу while
    print("No")
````

Но что, если первый раз код потребуется вызвать в начале, а потом только в конце, и при этом между ними будет много строчек? Тогда цикл не получится. На помощь приходят функции. Функция - это своего рода подпрограмма, которую можно вызвать в другой программе.

Функция в Питоне задается следующим образом:

```python
def имя_функции(параметры):
  тело функции
  return что-то
```

По умолчанию return равен None, его можно не писать.

````python
def function_name():
  pass  # Это слово в теле функции значит, что пока она не дописана и её можно пропустить при выполнении программы
````

````python
def clear_lower_letters(stroka):
  return ''.join(filter(lambda x: x.isupper(), stroka))

def clear_upper_letters(stroka):
  return ''.join(filter(lambda x: x.islower(), stroka))


string = "HElLo, i'm haPpy, i'M finE"
print(clear_lower_letters(string))
print(clear_upper_letters(string))
````

Причина в области видимости переменных. Тело главной программы не видит, что происходит в теле функции. Программа видит только то, что относится к ней самой. По этой же причине вы не можете обратиться извне к переменной, объявленной внутри цикла.

В Питоне по умолчанию все переменнные локальные. В некоторых языках, например в bash, все по умолчанию глобальные. Локальные - значит, обратиться к ним извне не получится. Глобальные - значит, их видно вообще ото всюду.

Правило простое: если вы хотите поменять переменную внутри функции - передайте её в качестве аргумента.

Если вам для каких-то целей все-таки потребуется обратиться к переменной главной программы из функции (обычно для глобального счетчика) - пропишите перед ней слово global. Это позволит изменять переменную на всех уровнях.

````python
def делайте_так():
  global внешняя_переменная
  внешняя_переменная += 10
  print(внешняя_переменная)


внешняя_переменная = 100
делайте_так()
````

Помимо global есть еще nonlocal. Оно используется, когда у нас функция-в-функции и мы хотим обратиться из внутренней функции ко внешней.

````python
def bar():
  var = 11
  def foo():
    nonlocal var
    var = 100
  foo()
  return var


var = 10
print(bar())
print(var)
````

def my_generator():

### Рекурсия

Рекурсивная функция - это функция, которая вызывает саму себя в своем теле. Это полезный подход при решении задач, которые могут быть разбиты на более мелкие подзадачи.

````python
def rec(i):
  if i >= 100:
    return i
  return rec(i + 5)


rec(3)
````

В качестве примера напишем программу рекурсивной проверки на палиндром.

````python
def palindrom(s):
  if len(s) <= 1:
    return True
  if s[0] != s[-1]:
    return False
  return palindrom(s[1:-1])
````

````python
print(palindrom("арозаупаланалапуазора"))
print(palindrom("urrf"))
````

Если мы пишем рекурсию, то всегда должно достигаться какое-либо состояние, когда из рекурсии можно выйти. То есть, рано или поздно функция должна достигнуть строчки return.

### Декораторы

Декораторы позволяют модифицировать поведение функций или классов без изменения их исходного кода. Они нужны для добавления дополнительной функциональности к существующим объектам.

* Функции в Python являются объектами, что означает, что их можно присваивать переменным, передавать в другие функции и возвращать из функций.
* Декораторы позволяют использовать функцию в качестве аргумента и обернуть ее в другую функцию.

````python
def burger_decorator(original_function):
    def wrapper_function(*args, **kwargs):
        # Добавляем начинку перед сборкой бургера
        print("Добавляем начинку")
        # Вызываем оригинальную функцию, которая собирает бургер
        result = original_function(*args, **kwargs)
        # Добавляем соус после сборки бургера
        print("Добавляем соус")
        # Возвращаем результат
        return result
    # Возвращаем обернутую функцию
    return wrapper_function


@burger_decorator  # Синтаксический сахар для "навешивания" декоратора
def make_burger():
    print("Собираем бургер")


# Вызываем декорированную функцию
make_burger()
````

## try-except

Это механизм для обработки исключений, которые могут возникать во время выполнения программы. Он позволяет перехватывать и обрабатывать исключения, чтобы предотвратить аварийное завершение программы.
Общий синтаксис:

```python
try:
    # Код, который может вызвать исключение
except ExceptionType1:
    # Обработка исключения типа ExceptionType1
except ExceptionType2:
    # Обработка исключения типа ExceptionType2
except (ExceptionType3, ExceptionType4):
    # Обработка исключения типа ExceptionType3 или ExceptionType4
else:
    # Код, который выполняется, если исключение не возникло
finally:
    # Код, который выполняется всегда, независимо от того, возникло исключение или нет
```

````python
try:
    x = 10 / 0  # Попытка деления на ноль
except ZeroDivisionError:
    print("Ошибка: деление на ноль!")
else:
    print("Результат:", x)
finally:
    print("Завершение программы")
````

*Опционально всё, кроме `try` и хотя бы одного `except`. В случае, если вы хотите просто подавить исключение, можно написать*

```python
try:
    # Ваш код
except:
    pass
```
