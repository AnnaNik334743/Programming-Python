## Списки

Список - изменяемый итерируемый тип данных. Обозначается словом list, его элементы записываются в квадратных скобках. Список - это набор различных элементов, причем необязательно одного типа. Изменяемый - значит можно поменять список частично без необходимости создавать новый объект. В других языках программирования похожий тип данных называется массивом с той оговоркой, что в массиве хранятся данные одного типа, а в списке - любого.

```python
some_list = [4, True, [0.0, None], 71, 'hello']
type(some_list)
```

```python
empty_list1 = []  # Как задать пустой список
empty_list2 = list()
print(empty_list1, empty_list2)
```

К элементам списка можно обращаться по индексу. В Python индексация с нуля (первый элемент имеет номер ноль).

```python
print(some_list[0])
print(some_list[-1])  # Обращение к последнему элементу
some_list[0] = 1999
print(some_list)
```

```python
print(some_list[10])  # Нельзя обратиться к несуществующему индексу
```

Рассмотрим некоторые основные действия над списками.

```python
empty_list = []
another_empty_list = list()

empty_list.append(1)  # Добавление элемента в конец списка
empty_list.append(8)
another_empty_list.append(100)

print(len(empty_list))  # Длина списка

print(empty_list + another_empty_list)  # Списки можно складывать
print(empty_list * 3)  # И умножать (список повторится несколько раз)

del empty_list[0]  # Можно удалить элемент списка, обратившись к нему по индексу
print(empty_list)
```

```python
1000 in another_empty_list  # Проверка, содержится ли элемент в списке
```

```python
array = [12, 54, 9, 87, 2]
print(min(array), max(array), sum(array))  # Только со списками, состоящими из чисел
```

Python имеет встроеннную сортировку. Обратите внимание: функция sorted() не изменяет сам список, а вот метод sort() - да.

```python
new_list = [1, 2, 7, 0, -100, - 20, 43]

print(sorted(new_list))
print(new_list)

new_list.sort()
print(new_list)
```

```python
sorted(new_list, reverse=True)  # Если нужно отсортировать по убыванию
```

Добавлять новые элементы в список можно по-разному. Результаты, соответсвенно, тоже будут разными.

```python
a = [1, 2, 3]
a.append([1, 2, 3])

b = [1, 2, 3]
b.extend([1, 2, 3])

print(a, b)
```

Еще несколько полезных методов и функций для работы со списками. Пусть наш список имеет имя `a`.

* `len(a)` - длина списка
* `a.append(x)` - добавляет элемент `x` в конец списка
* `a.extend(seq)` - добавляет в конец списка все элементы последовательности `seq`
* `del a[i]` - удаляет элемент с индексом `i`. Если такого элемента не окажется, выдаст ошибку
* `a.clear()` - делает список пустым
* `a.copy()` - создает копию списка
* `a.count(x)` - считает, сколько раз в списке встречается элемент `x`
* `a.index(x)` - выведет индекс первого найденного элемента со значением `x`
* `a.index(x, start, end)` - будет искать индекс элемента со значением `x` в диапазоне индексов от `start` до `end-1`
* `a.insert(i, x)` - вставит элемент `x` на место `i`, при этом все остальные элементы сдвинутся вправо
* `a.pop()` - удаляет последний элемент из списка, при этом возвращая его. В скобках можно указать индекс элемента, который нужно удалить и вывести
* `a.remove(x)` - удаляет первый найденнный элемент со значением `x`
* `a.reverse()` - разворачивает список в обратном направлении

Срезы - это когда мы берем не весь список, а только интересующую нас часть списка. `some_list[begin, end, step]` - возьмет значения списка из диапазона от `begin` до `end-1` с шагом `step`. Шаг по умолчанию равен одному.

```python
some_list = list(range(10))
print(some_list)
print(some_list[0:10:2])
print(some_list[8:4:-1])
print(some_list[:8])  # С самого начала
print(some_list[3:])  # И до самого конца
print(some_list[::])  # Весь список
print(some_list[:])  # Тоже весь список
print(some_list[::-1])  # Весь список задом-наперед
```

Для обхода списка можно использовать циклы.

```python
a = [1, 8, -2, 6, 3]
for i, elem in enumerate(a):
  print(f"Элемент {elem} стоит на четном месте" if i % 2 == 0 else f"Элемент {elem} стоит на нечетном месте")
```

Атомарные объекты (числа, строки) при их присваивании передаются по значению, а ссылочные (списки, кортежи, словари, функции, классы) — по ссылке.

Список является ссылочным типом данных. Поэтому одинаковые объекты ссылаются на одну и ту же ячейку памяти. Это буквально означает, что после присваивания вы имеете один-единственный объект, у которого несколько имен. Это может привести к следующему:

```python
a = [1, 2, 3]
b = a
a.append(8)  # Изменяем список a
print(a, b)  # Но изменились почему-то оба!
```

Чтобы избежать подобных ситуаций, можно воспользоваться одним из следующих способов:

```python
a = [1, 2, 3]
b = a.copy()  # shallow copy
a.append(8)
print(a, b)
```

```python
a = [1, 2, 3]
b = a[:]  # shallow copy
a.append(8)
print(a, b)
```

```python
from copy import deepcopy


a = [1, 2, 3]
b = deepcopy(a)
a.append(8)
print(a, b)
```

Аналогичная ситуация остальными ссылочными типами данных.

Начиная с версии 3.8 в Python появился следующий синтаксис. Оператор присвоения `:=` используется для вычисления и сохранения вычисленного значения, иногда это может быть удобно в случае работы со списковыми включениями или lambda-функциями.

```python
def f(x):
  return 2 * x


arr = [y for i in [1, -2, 3] if (y := f(i)) > 0]
arr
```

Поговорим о разных полезных методах для работы со строками. Пусть наша строка имеет имя `s`.

* `s.upper()` - меняет буквы нижнего регистра на верхний
* `s.lower()` - меняет буквы верхнего регистра на нижний
* `s.swapcase()` - меняет буквы верхнего регистра на нижний, а нижнего - на верхний
* `s.title()` - меняет первые буквы слов на заглавные, а все остальные на строчные
* `s.count('letters', begin, end)` - посчитает, сколько раз в строке встречается подстрока `letters`, где индексы принадлежат диапазону от `begin` до `end-1`. По умолчанию `begin` равен началу строки, а `end` - концу
* `s.find('e')` - ищет индекс первого вхождения символа `'e'` в строку `s`. Если ничего не найдено, вернет -1
* `s.rfind('e')` - аналогично предыдущему методу, но ищется индекс последнего вхождения
* `s.replace('a', 'b', n)` - заменит `n` первых символов `'a'` на `'b'`. Если не указать значение `n`, то заменятся все
* `s.isalpha()` - проверяет, состоит ли строка целиком из букв (не забываем про пробелы!) и возвращает `True` или `False`
* `s.isdigit()` - проверяет, состоит ли строка целиком из цифр
* `s.rjust(5, '0')` - дополнит слева исходную строку до 5 символов символами '0'
* `s.ljust(5, '0')` - дополнит справа исходную строку до 5 символов символами '0'
* `s.split('a')` - разобьет строку по указанным в скобках символам. По умолчанию разбивает по пробелам
* `len(s.split())` - позволяет в одну строчку узнать количество слов в строке
* `'x'.join(list)` - преобразует список `list` в строку, где `x` - любой разделитель между элементами списка
* `','.join(s.split())` - буквально замена одного разделителя другим
* `s.strip()` - удаляет из строки все пробелы и переносы (\n), стоящие вначале и конце строки
* `s.rstrip()` - то же самое, но только в конце
* `s.lstrip()` - то же самое, но только в начале

```python
s = """какая-то кривая строка
причем длинная    """
print('-'.join(s.split()))

print(len(s))
print(s.strip())
print(len(s.strip()))
```

Напишем небольшую программу, которая проверяет, является введенное пользователем число целым или вещественным. Напоминаю: Python воспринимает все, что введено с клавиатуры, как строки.

```python
s = input()
flag = True
if s[0] == "-":  # Если число начинается с минуса, его можно безболезненно отбросить, используя срезы
    s = s[1:]
k_points = 0
for i in s:
    if not i.isdigit():  # Если символ - не цифра
        if i == "." and k_points == 0:  # Вдруг пользователь поставил в числе миллион точек? непорядок
            k_points += 1
        else:
            flag = False
            break
if flag and len(s) > len(str(float(s))):  # Проверка на ведущие нули
    flag = False
print("Yes" if flag else "No")
```

Пример использования printable в задаче на проверку корректности введенного пользователем пароля. Хороший пароль должен быть не короче 16 символов и содержать:

* специсимволы
* буквы в разных регистрах
* числа

```python
from string import printable as pr


s = input()
if len(s) < 16:
    print('No')
else:
  flag = True
  symbols = [False, False, False, False]
  for i in s:
      if i in pr[:10]:
          symbols[0] = True
      elif i in pr[10:36]:
          symbols[1] = True
      elif i in pr[36:62]:
          symbols[2] = True
      elif i in pr[62:]:
          symbols[3] = True
  for j in symbols:
      if not j:
          print('No')
          flag = False
          break
  if flag:
    print('Yes')
```

## Кортежи

Кортежи похожи на списки, но они занимают меньше памяти и являются неизменяемыми. Их удобно использовать для хранения координат точек, передачи аргументов в функцию и для кортежного присваивания. К элементам кортежа можно обращаться по индексу.

```python
t = (1, 2, 3)
type(t)
```

```python
empty_tup = ()
empty_tup2 = tuple()
empty_tup == empty_tup2
```

```python
a, b, c = (12, 18, 6)  # Кортежное присваивание
print(a, b, c)
```

```python
seq = tuple('abcdef')
seq2 = list('abcdef')
print(seq, seq2)
```

Множества удобно использовать в задачах, где необходимо найти уникальные элементы. Например, пусть нам требуется посчитать, сколько раз символы встречаются в строке.

```python
def un_let(phrase: str):
  return sorted([(i, phrase.count(i)) for i in set(phrase)])


print(*un_let("я осуждал, осуждаю и буду осуждать"), sep='\n')
```

### JSON

JSON (JavaScript Object Notation) - это формат данных. JSON представляет данные в виде пар "ключ-значение" и поддерживает различные типы данных, такие как строки, числа, логические значения, массивы и объекты.

В Python есть встроенная поддержка для работы с JSON в виде стандартной библиотеки json, которая предоставляет функции для сериализации (конвертации объектов Python в JSON) и десериализации (конвертации JSON в объекты Python). Вообще, формат JSON - это буквально питоновские словари.

Сериализация (конвертация объектов Python в JSON):
Для сериализации объекта Python в JSON используется функция json.dumps(). Она принимает объект Python и возвращает его представление в виде строки JSON.

```python
import json


data = {
    'name': 'John',
    'age': 30,
    'city': 'New York'
}

json_data = json.dumps(data)
json_data
```

Для десериализации JSON в объекты Python используется функция json.loads(). Она принимает строку JSON и возвращает объект Python.

```python
json_data = '{"name": "John", "age": 30, "city": "New York"}'
data = json.loads(json_data)
print(data)
```

Чтение и запись данных в файлы JSON также является распространенной задачей. Для чтения JSON из файла используется метод json.load() и для записи JSON в файл используется метод json.dump().

```python
# Чтение JSON из файла
with open('data.json') as file:
    data = json.load(file)
    print(data)

# Запись JSON в файл
data = {
    'name': 'John',
    'age': 30,
    'city': 'New York'
}
with open('data.json', 'w') as file:
    json.dump(data, file)
```

## collections

Модуль `collections` - это встроенный модуль в Python, который предоставляет дополнительные структуры данных.

Counter (Счетчик):

Класс Counter используется для подсчета частоты элементов в итерируемом объекте. Он является подклассом словаря и принимает итерируемый объект в качестве входных данных. Возвращает словарь, в котором элементы являются ключами, а их частоты - значениями.

```python
from collections import Counter


# Подсчет частоты элементов в списке
fruits = ['яблоко', 'банан', 'апельсин', 'яблоко', 'банан', 'яблоко']
fruit_counts = Counter(fruits)
print(fruit_counts)

# Получение частоты определенного элемента
print(fruit_counts['яблоко'])
```

defaultdict (Словарь с заданным значением по умолчанию):

Класс defaultdict является подклассом словаря и предоставляет значение по умолчанию для отсутствующих ключей. При обращении к несуществующему ключу он создает этот ключ и присваивает значение по умолчанию, указанное при инициализации.

```python
from collections import defaultdict


# Создание defaultdict с значением 0 по умолчанию
num_counts = defaultdict(int)  # Можно передать lambda-функцию, генерирующую определенные значения

# Подсчет частоты чисел
numbers = [1, 2, 3, 1, 2, 3, 4, 5]
for num in numbers:
    num_counts[num] += 1

print(num_counts)
```

namedtuple (Именованный кортеж):

Класс namedtuple является подклассом кортежа и позволяет обращаться к элементам по именованным полям. Он предоставляет простой способ определения структуры, подобной классу, без написания кастомного класса.

```python
from collections import namedtuple


# Определение именованного кортежа для точки
Point = namedtuple('Point', ['x', 'y'])

# Создание объекта Point
p = Point(2, 3)
print(p.x, p.y)
```

deque (Двусторонняя очередь):

Класс deque представляет двустороннюю очередь, которая позволяет быстро вставлять и удалять элементы с обоих концов. Он предоставляет эффективный способ реализации очередей и стеков.

```python
from collections import deque


# Создание двусторонней очереди
d = deque([1, 2, 3])

# Добавление элементов справа
d.append(4)
d.append(5)

# Добавление элементов слева
d.appendleft(0)
d.appendleft(-1)

print(d)
```

OrderedDict (Упорядоченный словарь):

Класс OrderedDict является подклассом словаря и запоминает порядок вставки ключей. При итерации по словарю, он предоставляет предсказуемый порядок элементов.

```python
from collections import OrderedDict


# Создание упорядоченного словаря
d = OrderedDict()

# Добавление элементов в словарь
d['a'] = 1
d['b'] = 2
d['c'] = 3

# Печать ключей в порядке их добавления
for key in d:
    print(key)
```
